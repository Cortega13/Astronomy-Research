```python
import matplotlib.pyplot as plt

# Data for Timestep Multiplier = 10
dataset_fraction_10 = [0.5, 0.6, 0.7, 0.8, 0.9, 1.0]
average_error_10 = [1.3766e+06, 1.3228e+06, 1.1346e+06, 1.0241e+06, 1.0057e+06, 1.1853e+06]

# Data for Timestep Multiplier = 1
dataset_fraction_1 = [0.5, 0.6, 0.7, 0.8, 0.9, 1.0]
average_error_1 = [1.8580e+02, 2.3579e+02, 2.5479e+02, 3.2906e+02, 1.9317e+02, 3.4002e+02]

# Plot for Timestep Multiplier = 10
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(dataset_fraction_10, average_error_10, marker='o', label='Evolution = 10,000y')
plt.xlabel('Dataset Fraction')
plt.ylabel('Average Error')
plt.title('Average Error vs Dataset Fraction (Evolution = 10,000y)')
plt.grid(True)
plt.legend()

# Plot for Timestep Multiplier = 1
plt.subplot(1, 2, 2)
plt.plot(dataset_fraction_1, average_error_1, marker='o', color='red', label='Evolution = 1000y')
plt.xlabel('Dataset Fraction')
plt.ylabel('Average Error')
plt.title('Average Error vs Dataset Fraction (Evolution = 1000y)')
plt.grid(True)
plt.legend()

# Show plots
plt.tight_layout()
plt.show()
```
